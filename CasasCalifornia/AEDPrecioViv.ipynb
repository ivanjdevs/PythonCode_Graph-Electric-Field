##### La siguiente jugada consiste en armar una pequeña tabla, tomando del data frame principal las columnas 'SalePrice' y  'GrLivArea'. Axis=1 significa para Python que la concatenación la haga a través de las columnas o más bien, en el eje vertical

```python
tabla1=pd.concat([df_train['SalePrice'],df_train['GrLivArea']], axis=1)
```

```python
tabla1.head(20)
```

##### Ahora con la tabla creada se hará un gráfica scatterplot

```python
tabla1.plot.scatter(x='GrLivArea', y='SalePrice', ylim=(0,800000))
```

##### Se grafica ahora el Área del sótano vs el Precio de venta

```python
tabla2=pd.concat([df_train['SalePrice'],df_train['TotalBsmtSF']], axis=1)
tabla2.plot.scatter(x='TotalBsmtSF', y='SalePrice', ylim=(0,800000))
```

### Relaciones categóricas. Precio de venta vs. Calidad

```python
tabla3=pd.concat([df_train['SalePrice'],df_train['OverallQual']], axis=1)
tabla3.head(10)
```

#### Ahora, en dos variables f y ax, vamos a guardar el resultado de un plt.subplots;

```python
f, ax=plt.subplots(figsize=(8,6)) #definir el área del gráfico o pedazo de pantalla que reservo para la gráfica (de la librería matplotlib vas a agarrar subplots)
fig=sns.boxplot(x='OverallQual', y='SalePrice', data=tabla3)  # definir la propia gráfica
fig.axis(ymin=0, ymax=800000) #modificando el atributo axis de la figura definida fig
```

#### Matriz de correlación
##### Permite ver que variables están mas correlacionadas

```python
#Definir el nombre dela matriz
MatCorrel=df_train.corr() #el método .corr permite hallar la correlación entre todas las variables del dataframe
```

```python
MatCorrel.head(5)
```

```python
f, ax=plt.subplots(figsize=(12,9))
sns.heatmap(data=MatCorrel, vmax=0.8, square=True)
```

#### Matriz de correlación con números

```python
k=10
```

```python
#agárrate ahi de la matriz de correlación creada, las k variables que están mas correlacionadas con Saleprice
cols=MatCorrel.nlargest(k, 'SalePrice')['SalePrice'].index  
```

```python
cols
```

```python
cm=np.corrcoef(df_train[cols].values.T)
```

```python
cm
```

```python
sns.set(font_scale=1.25)  #el tamaño de la fuente de los números del mapa de calor que se va a dibujar enseguida, sea de 1.25
```

```python
hm=sns.heatmap(cm, cbar=True, annot=True, square=True, fmt=".2f", annot_kws={"size":10,}, yticklabels=cols.values, xticklabels=cols.values)
plt.show
```
